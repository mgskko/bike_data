```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'

# Excel 파일 읽기
df = pd.read_csv('공공자전거 대여소 정보(22.12월 기준).csv')
```

```python
df
```

```python
df.info()
```

```python
# nan 값 0으로 변환
df.fillna(0, inplace = True)
```

```python
# LCD, QR int형으로 변환
df['LCD'] = df['LCD'].astype(int)
df['QR'] = df['QR'].astype(int)
```

```python
df.info()
```

```python
# 자치구별 따릉이 대여소 수
df['자치구'].value_counts()
```

```python
# 기본 데이터 설정
count = df['자치구'].value_counts()

# 그래프 크기 설정
plt.figure(figsize=(10, 5))

# 막대그래프 생성 및 설정
bars = plt.bar(count.index, count, align='center', edgecolor='lightgray', color='#1BCF1C')

# x축과 y축 레이블 설정
plt.xlabel('자치구')
plt.ylabel('따릉이 대여 장소 수')

# x축 라벨 회전 및 그리드 설정
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 그래프 제목 설정 및 위치 조절
plt.title('<자치구별 따릉이 대여 장소 수>', fontsize = 20, pad=20)

# y축 범위 조절
plt.ylim(0, max(count)*1.1)

# 각 막대에 숫자 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 2), ha='center', va='bottom')

# 그래프 표시
plt.tight_layout()  # 그래프의 레이아웃을 조정하여 라벨, 제목 등이 잘리지 않게 함
plt.show()
```

```python
df['총 따릉이 수'] = df['LCD'] + df['QR']
```

```python
# 자치구별 '총 따릉이 수'를 계산하고 내림차순으로 정렬
sorted_count = df.groupby('자치구')['총 따릉이 수'].sum().sort_values(ascending=False)

# 가장 많은 따릉이가 있는 자치구의 인덱스 찾기
max_value_index = sorted_count.index[0]

# 그래프 크기 설정
plt.figure(figsize=(10, 5))

# 가장 높은 값의 자치구만 빨간색으로 설정, 나머지는 초록색으로 설정
colors = ['#1BCF1C' if idx != max_value_index else 'red' for idx in sorted_count.index]

# 막대그래프 생성 및 설정
bars = plt.bar(sorted_count.index, sorted_count, align='center', edgecolor='lightgray', color=colors)

# x축과 y축 레이블 설정
plt.xlabel("자치구", fontsize=12)
plt.ylabel("자치구별 자전거 수", fontsize=12)

# x축 라벨 회전 및 그리드 설정
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# y축 범위 조절
plt.ylim(0, max(sorted_count)*1.1)

# 각 막대에 숫자 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 2), ha='center', va='bottom', fontsize=9)  # fontsize=9 으로 설정

# 제목 설정
plt.title('<자치구별 총 따릉이 수>', fontsize=20, pad=20)

# 그래프 표시
plt.tight_layout()  # 그래프의 레이아웃을 조정하여 라벨, 제목 등이 잘리지 않게 함
plt.show()
```
